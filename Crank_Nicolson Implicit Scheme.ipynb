```python
import math   ## importing the required libraries
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
c = 910
p = 2699
k = 205.0
x_0 = 1 # natural length scale
t_0 = (c * p * (x_0)**2)/k # natural time scale
T_0 = 273 # natural temperature scale
tn=0 # dimensionless time
X=0 # dimensionless length
n=1000 # no. of divisions on time scale(original)
m=10 # no. of divisions on length scale(original)
h=1/(m)
k=1/(n)
l= k/(h**2)
print(l)
iter=100
```

```python
matA=np.zeros((n,m))
matC =np.zeros((n,m))
for i in range (0,n):
    for j in range (0,m):
        matC[i][j]=1
        matA[i][j]=1
for i in range (0,m):
    matA[0][i]=373/T_0
for i in range (1,n):
    matA[i][0]= 273/T_0
    matA[i][m-1]= 273/T_0
for steps in range(0,iter):
    for i in range (0,n-1):
        for j in range (1,m-1):
            matC[i][j] = l*(matA[i][j-1])+ 2*(1-l)*(matA[i][j]) + l*(matA[i][j+1])
            matA[i+1][j]= (l/(2*(1+l)))*(matA[i+1][j-1] + matA[i+1][j+1] ) + (1/(2*(1+l)))*matC[i][j]
print(matA)
```

```python
Xlist=[]
for i in range (0,m+1):
    Xlist.append(X)
    X=X+h
matB= np.zeros((n,m))
for i in range (0,n):
    for j in range (0,m):
        matB[i][j]= 273*matA[i][j]
```

```python
for i in range (0,10):
    plt.plot(matB[10*i,:])
plt.xlabel('length in metres')
plt.ylabel('tempersture in kelvins')
```
