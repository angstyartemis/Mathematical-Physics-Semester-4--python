```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import cmath
```

```python
N=1000
xi=[2*math.pi*i/N for i in range (0,N)]
yi=[math.sin(p) for p in xi]
C_k_squared=[]
k_list=[]
C_list=[]
for k in range (0,N):
    s=complex(0,0)
    for n in range (0,N):
        s = s + yi[n]*cmath.exp((-2*math.pi*complex(0,1)*k*n)/N)
    C= complex(s/N)
    C_k2=abs(C)**2
    C_list.append(C)
    C_k_squared.append(C_k2)
    k_list.append(k)
```

```python
plt.plot(k_list, C_k_squared)
plt.title('|C_k|^2 vs k graph')
plt.xlabel('k')
plt.ylabel('|C_k|^2')
```

```python
y_n_list=[]
for n in range (0,N):
    p=complex(0,0)
    for k in range (0,N):
        p = p + C_list[k]*cmath.exp((2*math.pi*complex(0,1)*k*n)/N)
    y_n_list.append(p.real)
```

```python
plt.plot(k_list, y_n_list)
plt.title('y_n vs k graph (f(x)=sinx) ')
plt.xlabel('k')
plt.ylabel('y_n')
```
