```python
from mpl_toolkits import mplot3d
import math   ## importing the required libraries
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=0
y=0
n=100
h=1/n
it=100
```

```python
matA=np.zeros((n,n))
for i in range (0,n):
    for j in range (0, n):
        matA[i,j]=1
for i in range (0,n):
    matA[0,i]= 0
    matA[-1,i]= 100
    matA[i,0]=0
    matA[i,-1]=0
print(matA)
```

```python
for steps in range(0,it):
    for i in range (1,n-1):
        for j in range (1,n-1):
            matA[i,j]= (matA[i+1,j] + matA[i-1,j] + matA[i,j+1] + matA[i,j-1])/4
print(matA)
```

```python
X=[]
Y=[]
Z=np.zeros((n,n))
for i in range (0,n):
    X.append(x)
    Y.append(y)
    x=x+h
    y=y+h
for i in range (0,n):
    for j in range (0,n):
        Z[i,j]=matA[i,j]
```

```python
fig = plt.figure()
ax=plt.axes(projection='3d')
X,Y = np.meshgrid(X,Y)
ax.contour3D(X,Y,Z, 100, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u')
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z,1000,cmap='RdGy')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u');
```
