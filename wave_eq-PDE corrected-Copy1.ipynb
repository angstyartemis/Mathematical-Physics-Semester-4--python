```python
l=float(input('enter the value of the length of string:'))
```

```python
tau=0
X=0
n=150 ## time steps
m=50 ## x steps
k=1/n
h=1/m
lam= k/h
it=100
Xs=[]
taus=[]
Z=np.zeros((m+1,m+1))
for i in range (0,m+1):
    Xs.append(X)
    X=X+h

for i in range (0,m+1):
    taus.append(tau)
    tau=tau+k
    
print(Xs)
print(taus)
print(lam)
```

```python
Y=np.zeros((m+1,m+1))
for i in range (0,m+1):
    for j in range (0, m+1):
        Y[i,j]=0

for i in range (1,m):
    if( Xs[i]< 0.5):
        Y[0,i]=2*l*Xs[i]
    else:
        Y[0,i]=2*l*(1-Xs[i])
        
for i in range (1,m+1):
    Y[i,0]= 0
    Y[i,-1]= 0
#print(Y)
for steps in range(0,it):
    for j in range (0,m):
        for i in range (1,m):
            if(j==0):
                Y[j+1,i]= (lam**2)*(Y[j,i+1]) + (1-lam**2)*(Y[j,i]) + (lam**2)*(Y[j,i-1]) - Y[j,i]
            else:
                Y[j+1,i]= (lam**2)*(Y[j,i+1]) + (1-lam**2)*(Y[j,i]) + (lam**2)*(Y[j,i-1]) - Y[j-1,i]
print(Y)
```

```python
import matplotlib.pyplot as plt 
plt.figure(dpi = 100)
for i in range(0, 50):
    plt.plot(Xs, Y[i][:], label = 't = {}'.format(i))
plt.legend();


'''fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.contour3D(taus,X,Z,1000,cmap='RdGy')
ax.set_xlabel('x')
ax.set_ylabel('taus')
ax.set_zlabel('y');'''
```
